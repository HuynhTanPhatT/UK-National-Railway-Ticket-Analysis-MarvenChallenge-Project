## Data Description

- **`Transaction ID`**: Unique identifier for an individual train ticket purchase
- **`Date of Purchase`**: Date the ticket was purchased
- **`Time of Purchase`**: Time the ticket was purchased
- **`Purchase Type`**: Whether the ticket was purchased online or directly at a train station
- **`Payment Method`**: Payment method used to purchase the ticket (Contactles Credit Card or Debit Card)
- **`Railcard`**: Whether the passenger is a National Railcard holder (Adult Senior or Disabled) or not (None). Railcard holders get 1/3 off their ticket purchases.
- **`Ticket Class`**: Seat class for the ticket (Standard or First)
- **`Ticket Type`**: When you bought or can use the ticket.
    - **Advance tickets** are 1/2 off and must be purchased at least a day prior to departure. 
    - **Off-Peak tickets** are 1/4 off and must be used outside of peak hours - (weekdays between 6-8am and 4-6pm).
    - **Anytime tickets** are full price and can be bought and used at any time during the day.
- **`Price`**: Final cost of the ticket
- **`Departure Station`**: Station to board the train
- **`Arrival Destination`**: Station to exit the train
- **`Date of Journey`**: Date the train departed
- **`Departure Time`**: Time the train departed
- **`Arrival Time`**: Time the train was scheduled to arrive at its destination (can be on the day after departure)
- **`Actual Arrival Time`**: Time the train arrived at its destination (can be on the day after departure)
- **`Journey Status`**: Whether the train was on time delayed or cancelled
- **`Reason for Delay`**: Reason for the delay or cancellation
- **`Refund Request`**: Whether the passenger requested a refund after a delay or cancellation

## Data Quality Report

```python
def check_data_quality(df):
    quality_report = {
        'null_values': df.isnull().sum().to_dict(),
        'duplicate_values': df.duplicated().sum(),
        'info_values': df.info()
    }
    return quality_report
```

```python
display(check_data_quality(railway_ticket))
```

```python
# Percentage of null values
pct_null_values = (railway_ticket.isnull().sum() * 100 / len(railway_ticket)).sort_values(ascending=False)
pct_null_values.loc[pct_null_values > 0]
```

Let's take a look:
- The dataset includes **31,652** rows and **18** columns. There are no duplicate ticket transactions, but several columns contains **null values**:
    - `Reason for Delay` and `Railcard` have the highest missing values (more than **50%**).
    - `Actual Arrival Time` has **1,880** rows accounting for **6%** of the dataset.

## Handling Null Values

```python
# "Reason for Delay" column
railway_ticket['Reason for Delay'] = np.where(railway_ticket["Reason for Delay"].isnull(),"None",railway_ticket["Reason for Delay"])
```

```python
# "Railcard" column
railway_ticket['Railcard'] = np.where(railway_ticket["Railcard"].isnull(),"None", railway_ticket["Railcard"])
```

- `Reason for Delay` is null because the train journey was not delayed, therefore there is no reason for recorded.
- `Railcard` is null because these passenger are not **National Railcard holders**.

### B. "arrival_date" column 

Identified **916** cases where the `Actual Arrival Time` < the `Departure Time`, these are **overnight train**

```python
railway_ticket.loc[:,["Date of Journey", "Departure Time", "Actual Arrival Time"]]
```

Conditions:
- If **Actual Arrival Time** > **Departure Time** => Keep the `Date of Journey` as the `Date of Arrival`
- If **Actual Arrival Time** < **Departure Time** => `Date of Arrival` = `Date of Journey` + 1 day

```python
railway_ticket.loc[railway_ticket["Actual Arrival Time"] < railway_ticket["Departure Time"],:].head()
```

```python
# Äiá»u kiá»‡n
condition = railway_ticket["Departure Time"] > railway_ticket["Actual Arrival Time"]

# Nháº­p dá»¯ liá»‡u Ä‘á»ƒ Arrival = Journey, sau Ä‘Ã³ má»›i báº¯t Ä‘áº§u +1 ngÃ y vá»›i cÃ¡c rows thá»a Ä‘iá»u kiá»‡n
railway_ticket["Date of Arrival"] = railway_ticket["Date of Journey"]

# Náº¿u thá»a Ä‘iá»u kiá»‡n = + thÃªm 1 ngÃ y => Arrival Date = Journey + 1 day
railway_ticket.loc[condition,"Date of Arrival"] = railway_ticket.loc[condition,"Date of Journey"] + pd.Timedelta(days=1)
```

```python
railway_ticket.loc[condition,:].head()
```

=> Expected Result: after applying this logic, each rows guarantee that: `Departure Time` < `Actual Arrival Time` => Logical timeline
